```python
import pandas as pd
import matplotlib.pyplot as plt

dtf = pd.read_csv('db_fifa.csv', low_memory=False,delimiter=';', encoding='latin-1', decimal=',')

dtf.columns
```

```python
nacionalidade = list(dtf['nationality'].value_counts().index.tolist())
idade = list(dtf['age'].value_counts().index.tolist())
altura = list(dtf['height'].value_counts().index.tolist())
peso = list(dtf['weight'].value_counts().index.tolist())
posicao = list(dtf['position'].value_counts().index.tolist())

peso.sort(key=int, reverse=True)
idade.sort(key=int, reverse=True)
altura.sort(key=int, reverse=True)
```

```python
idade = list(dtf.age[-10:-1].sort_index())
piores = list(dtf.player_name[-10:-1].sort_index())

plt.plot(idade,piores,'*r')
plt.plot(idade,piores, 'b-', color='green')

plt.title("Piores jogadores e suas idades (overall)")

plt.grid(False)

plt.show()
```

```python
p1 = []
p2 = []
x = 0
while x < 17:
    p1.append(dtf['position'].value_counts().values[x].tolist())
    p2.append(dtf['position'].value_counts().index[x])
    x = x+1

plt.title("Número de jogadores por posição")

plt.figure(1, figsize=(45,85))

plt.barh(p2, p1)

plt.show()
```

```python
overall = list(dtf.overall[0:10].sort_index())
melhores = list(dtf.player_name[0:10].sort_index())

plt.plot(overall,melhores,'ro')
plt.plot(overall,melhores, 'k-', color='green')

plt.title("Melhores jogadores e suas pontuações (overall)")

plt.grid(True)

plt.show()
```
